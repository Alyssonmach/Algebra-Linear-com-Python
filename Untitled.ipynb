```python
R = img[:,:,0]
G = img[:,:,1]
B = img[:,:,2]
```

```python
img_BGR = np.array([B, G, R])
img_BGR = np.swapaxes(img_BGR, 0, 1)
img_BGR = np.swapaxes(img_BGR, 1, 2)
```

```python
plt.imshow(img_BGR)
plt.title("imagem BGR")
```

```python
img_BGR = np.array([np.invert(B), np.invert(G), np.invert(R)])
img_BGR = np.swapaxes(img_BGR, 0, 1)
img_BGR = np.swapaxes(img_BGR, 1, 2)
img_BGR.shape
```

```python
plt.imshow(img_BGR)
```

```python
R = img[:,:,0]
G = img[:,:,1]
B = img[:,:,2]
```

```python
img_BGR = np.array([np.transpose(B), np.transpose(G), np.transpose(R)])
img_BGR = np.swapaxes(img_BGR, 0, 1)
img_BGR = np.swapaxes(img_BGR, 1, 2)
img_BGR.shape
```

```python
plt.imshow(img_BGR)
```
